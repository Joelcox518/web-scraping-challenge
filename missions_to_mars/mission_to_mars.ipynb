```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

*NASA Mars News*


```python
#scrape website
url= 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'

#retrieve page data
html=requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
soup
```

```python
#pulling title
news_title= soup.find_all('div', class_='content_title')
news_title


for title in news_title:
    
    title_line=title.find('a')
    title=title_line.text
    print(title)
```

```python
#pulling paragraph text

news_p= soup.find_all('div', class_='rollover_description_inner')
news_p

for p in news_p:
    pgraph= p.text
    
    print(pgraph)
```

*Splinter*

```python
#splinter exercise
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url= 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)
```

FEATURED IMAGE

```python
# HTML object
html = browser.html
# Parse HTML with Beautiful Soup
soup2 = BeautifulSoup(html, 'html.parser')
#print(soup2)
    
img= soup.find(class_='carousel_item')['style']
#print(img)

string= img.split("'")[1]
#print(string)

base_url= 'https://jpl.nasa.gov'

featured_image_url= base_url + string
print(featured_image_url)
```

```python
browser.quit()
```
